```python
# Remove columns that contain only missing values
# The previously listed columns were not found in the DataFrame, so they have been removed from the drop list.

print("Unnecessary columns removed successfully.")
```

```python
# Remove columns that contain only missing values
duplicate_count = df.duplicated().sum()

print("Duplicate Rows:", duplicate_count)
```

```python
# Clean text columns
# Remove leading/trailing spaces
text_columns = df.select_dtypes(include='object').columns

for col in text_columns:

    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)

print("Text columns cleaned successfully.")
```

```python
# Convert data types
df['date'] = pd.to_datetime(df['date'])
integer_columns = [
    'match_id',
    'innings',
    'over',
    'ball',
    'ball_no',
    'bat_pos',
    'day',
    'month',
    'year'
]

for col in integer_columns:
    df[col] = df[col].fillna(0).astype(int)

print("Data types updated successfully.")
```

```python
# Standardize column names
df.columns = (df.columns
                .str.strip()
                .str.lower()
                .str.replace(" ", "_"))

print("Column names standardized successfully.")
```

```python
# Create a new column: ball_result
def ball_result(runs):
    if runs == 0:
        return "Dot Ball"
    elif runs == 1:
        return "Single"
    elif runs == 2:
        return "Double"
    elif runs == 3:
        return "Triple"
    elif runs == 4:
        return "Four"
    elif runs == 6:
        return "Six"
    else:
        return "Other"

df["ball_result"] = df["runs_batter"].apply(ball_result)

print("New column 'ball_result' created successfully.")
```

```python
# Display the first few rows of the new column
df[["runs_batter", "ball_result"]].head(10)
```

```python
# Final dataset information
print("Dataset Shape:", df.shape)
print("\nMissing Values:\n")
print(df.isnull().sum())

print("\nDataset Information:\n")
df.info()
```

```python
# Export cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset exported successfully as 'cleaned_dataset.csv'")
```
